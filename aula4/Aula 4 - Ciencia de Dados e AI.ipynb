```python
import matplotlib.pyplot as plt #exibir
import  seaborn as sns #criar

# correlacao das informacoes, explorar como as coisas estao acontecendo
print(tabela.corr())
#criar um grafico e exibir
sns.heatmap(tabela.corr(), cmap="Blues", annot=True)
plt.show()

```

```python
#criacao da AI - modelo de previsao

y= tabela["Vendas"]
x= tabela[["TV", "Radio", "Jornal"]] #quando for mais de uma coluna, colocar entre dois colchetes, pq e uma lista

from sklearn.model_selection import train_test_split #separando

x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size=0.3, random_state=1)

```

```python
#importar a AI
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

#criar
modelo_regressaolinear = LinearRegression()
modelo_arvoredecisao =RandomForestRegressor()

#treinar
modelo_regressaolinear.fit (x_treino, y_treino)
modelo_arvoredecisao.fit (x_treino, y_treino)
```

```python
previsao_regressaolinear = modelo_regressaolinear.predict (x_teste) 
previsao_arvoredecisao =  modelo_arvoredecisao.predict (x_teste)

from sklearn.metrics import r2_score #calculando o R2 para a previsao

print (r2_score(y_teste, previsao_regressaolinear))
print (r2_score(y_teste, previsao_arvoredecisao))
```

```python
#grafica das previsoes
tabela_auxiliar =pd.DataFrame () #cria uma tabela do 0.
tabela_auxiliar ["y_teste"] = y_teste #adicionando colunas
tabela_auxiliar ["Previsao Arvore Decisao"] = previsao_arvoredecisao
tabela_auxiliar ["Previsao Regressao Linear"] = previsao_regressaolinear
display (tabela_auxiliar)

sns.lineplot (data=tabela_auxiliar)
plt.show()
```
